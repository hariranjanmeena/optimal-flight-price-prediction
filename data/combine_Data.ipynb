```python
import pandas as pd
```

```python
import os
```

```python
Classes = os.listdir('data2/')
```

```python
Classes
```

```python
for folder in Classes:
    files = os.listdir(f'Data/{folder}')
    for f in files:
        df = pd.read_csv(f'data/{folder}/{f}')
        df['avg_Price'] = df['Price'].mean()
        df['min_Price'] = df['Price'].min()
        df.to_csv(f'data2/{folder}/{f}', index = False)
    print("pass",folder)
```

```python
for folder in Classes:
    # List of Files in old Folder
    files = os.listdir(f'data2/{folder}')
    print("pass",folder)
    combined_csv = pd.concat([pd.read_csv(f'data2/{folder}/{f}') for f in files ])
    combined_csv.to_excel( f"combined_Data/{folder}.xls", index=False, encoding='utf-8-sig')
    
        
```
